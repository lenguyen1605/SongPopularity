```python
# Popularity and Audio features
selected_columns = ['instrumentalness', 'danceability', 'liveness', 'energy', 'acousticness', 'valence', 'tempo', 'duration_ms']
correlation_table = pd.DataFrame(columns=['Audio Feature', 'Correlation'])
for feature in selected_columns:
    correlation = df['popularity'].corr(df[feature])
    correlation_table.loc[len(correlation_table)] = [feature, correlation]
correlation_table
```

We want to address the collinearity issues in our dataset. First, we create a correlation matrix between the predictors.

```python
# Convert the explicit predictor to 0 and 1 (true and false)
df['explicit'] = df['explicit'].astype(int)
df
```

```python
# Select quantitative predictors:
quantitative_predictors = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = quantitative_predictors.corr()
correlation_matrix
```

From this matrix, we observe a high positive correlation between danceability and loudness (0.7588), which means that the louder the audio, the higher the danceability of the music. Moreover, the strong negative correlation between energy and acousticness is noticeable (-0.7326). 
Thus, we will combine the loudness and dancebality to 1 new feature called 'loud_dance', which is the average of both predictors; and removinng acousticness from our dataset, as it can be inferred from the energy feature.

```python
df['loud_dance'] = (df['danceability'] + df['loudness']) / 2
df
```

Split the data into training and testing datasets

```python
# Drop loudness and danceability due to the new combined predictor loud_dance
df = df.drop('loudness', axis=1)
df = df.drop('danceability', axis=1)
df
```

```python
# Drop acousticness due to its high correlation with energy
df = df.drop('acousticness', axis=1)
df
```

The popularity of a song can be related to the number of artists participated in it. For example, 2 artists collaborating might attract more listeners, as each has their own fans. Therefore, we create a new predictor that represents the number of artists on this track.

```python
# The number of artists featured on a track can be represented by the number of occurences of ";"
df['num_artists'] = df['artists'].str.count(';') + 1
df
```
